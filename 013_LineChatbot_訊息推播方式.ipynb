```python
"""
訊息的推播方式： (透過業務傳訊息給用戶)

這邊只提兩種
1. 單一推播     => line_bot_api.push_message("用戶的ID", [傳送的訊息(最多可以傳五則)])

2. 指定用戶群推播  => line_bot_api.multicast("用戶群ID陣列", [傳送的訊息(最多可以傳五則)])
  - 先載入蒐集用戶個資的文件
  - 將每一位用戶的個資訊息轉換成json格式，並存放到一個array中
  - 然後再從json的array中，提取每位用戶的ID，並存放到一個新的user_id_array
  - 將推播訊息傳送給user_id_array中的每位用戶

"""

# 單一推播 push_message("用戶ID", [傳送訊息xxxxxSendMessage, xxxxxSendMessage, .....最多五則])
line_bot_api.push_message('U3741b2fcdd172578b072465a8b6a8fa3', [TextSendMessage(text="單一推播測試1"), TextSendMessage(text="單一推播測試2")])


##############################################################################################################################


# 指定用戶群推播 line_bot_api.multicast(用戶群ID陣列, [傳送訊息xxxxxSendMessage, xxxxxSendMessage, .....最多五則])

import json


# 1. 先載入蒐集用戶個資的文件
users = open("/content/drive/MyDrive/users.txt", "r", encoding="utf8")

# 2. 將每位用戶的個資訊息轉換成json格式並存放到一個新的array中
user_array = []

for user in users:
  user_json = json.loads(user)
  user_array.append(user_json)

# 3. 從新的array中提取每位用戶的ID，再存放到一個新的array中
users_id = []

for user in user_array:
  id = user.get("user_id")
  users_id.append(id)

# 4. 推播訊息給用戶群
line_bot_api.multicast(users_id, [TextSendMessage(text="群播測試1"), TextSendMessage(text="群播測試2")])
```
