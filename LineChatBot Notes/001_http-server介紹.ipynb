```
"""
安裝完套件後

再來 => 引用套件
"""
# 引用Flask Web Server套件
from flask import Flask, request, abort, jsonify

# 引用json套件
import json

# 引用flask-ngrok，自動生成外部連結的套件
from flask_ngrok import run_with_ngrok
```

```
"""
安裝完套件
引用完套件
再來 => 建置主程序
"""

# 設定Flask Web Server啟用細節

app = Flask(__name__, static_url_path = "/material", static_folder = "./material/")
"""
Flask()的參數說明：
__name__ 固定用法
static_url_path = "要移除的舊有路徑或網址(即移除掉原本路徑)" (i.e. the default /static).
static_folder = "靜態文件提供的資料夾路徑(即從哪邊讀取到文件，要設定新路徑)"      
"""

run_with_ngrok(app)
```

```
"""
安裝套件
引用套件
建置主程序

再來 => 設置主程序的多個API接口 (注意使用的方法)
"""

# 製作啟動主程序的API接口

# 創立並啟動主程序的第一個API接口(最簡單):
@app.route("/", methods=["GET"])
def callback():
  return "Hello World!"
  # 臨時網址/  =>  頁面會顯示Hello World!
```

```
# 再創立並啟動一個主程序的API接口名為calculate (計算機，觀察並學會剖析參數):
@app.route("/calculate", methods=["GET"])
def calculate():
  first_number = int(request.args.get("n1"))
  second_number = int(request.args.get("n2"))
  return str(first_number + second_number)
  # 臨時網址/calculate?n1=3&n2=5  =>  頁面會顯示8
```

```
"""
安裝套件
引用套件
建置主程序
建立並啟動主程序的API接口

最後 => 啟動主程序
"""

# 注意，啟動後會持續運轉，此cell以下的內容將無法運作
# 另外，若要修改任何東西請務必要先停止此cell，然後再重新跑過修改的cell，才能再運行程序，否則修改無效
app.run()
```
